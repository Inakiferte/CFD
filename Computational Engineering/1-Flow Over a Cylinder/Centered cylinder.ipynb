```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D


L = 10.0
H = 10.0
points = 100                             # Quantity of x points
D      = 2.0                             # Radius
X      = np.linspace(-D/2.0, D/2.0, 100) # Generate X points with the same spacing
Y      = np.zeros(100)                   # Generate Y vector
x0 = L / 2.0
y0 = H / 2.0
R0 = np.sqrt((L / 2.0)**2 + (H/2.0)**2)
# Fill Y vector



for i in range(100):
    
    Y[ i ] = np.sqrt((D / 2.0)**2 - X[ i ] ** 2)

fontsize = 30
plt.figure(figsize = (8,8))
plt.xlabel(r"$\frac{x}{R}$", fontsize=fontsize)
plt.ylabel(r"$\frac{y}{R}$" , fontsize=fontsize)

# Plot the points
plt.plot(X + x0, Y + y0, color = "black")
plt.plot(X + x0,-Y + y0, color = "black");
```
